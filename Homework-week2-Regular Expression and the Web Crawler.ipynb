# 正则表达式与爬虫

我们在这里回顾一些正则表达式的基础知识，并且给出简单的应用。

首先回忆课堂中的例子，用正则表达式爬去豆瓣电影的图片。

```python
import re
# 想要抓取网页可以用这个包
import requests
# 获取目标链接
url = 'https://movie.douban.com/'
response = requests.get(url)
```

```python
# re.compile() 函数根据包含的正则表达式的字符串创建模式对象
image_pattern = re.compile('https://img3.doubanio.com/view/photo/s_ratio_poster/public/\w\d+.\w+')

# 把网页内容转化为文本
html_content = response.text

# 找到所有的符合条件的图片地址
images = image_pattern.findall(html_content)
```

```python
# 显示图片的一种方法
from skimage import io
for i in range(len(images)):
    image = io.imread(images[i])
    io.imshow(image)
    io.show()
```
