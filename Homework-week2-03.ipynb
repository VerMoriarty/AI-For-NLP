```python
# 利用网上的代码把百科的内容中无用的符号去掉

import logging,jieba,os,re
 
def get_stopwords():
    logging.basicConfig(format='%(asctime)s:%(levelname)s:%(message)s',level=logging.INFO)
    #加载停用词表
    stopword_set = set()
    with open("C:/Users/cam/NLP/NLP_Week2/wikiextractor/extracted/AA/stopwords.txt",'r',encoding="utf-8") as stopwords:
        for stopword in stopwords:
            stopword_set.add(stopword.strip("\n"))
    return stopword_set
 
'''
使用正则表达式解析文本
'''
def parse_zhwiki(read_file_path,save_file_path):
    #过滤掉<doc>
    regex_str = "[^<doc.*>$]|[^</doc>$]"
    file = open(read_file_path,"r",encoding="utf-8")
    #写文件
    output = open(save_file_path,"w+",encoding="utf-8")
    content_line = file.readline()
    #获取停用词表
    stopwords = get_stopwords()
     #定义一个字符串变量，表示一篇文章的分词结果
    article_contents = ""
    while content_line:
        match_obj = re.match(regex_str,content_line)
        content_line = content_line.strip("\n")
        if len(content_line) > 0:
            if match_obj:
                #使用jieba进行分词
                words = jieba.cut(content_line,cut_all=False)
                for word in words:
                    if word not in stopwords:
                        article_contents += word+" "
            else:
                if len(article_contents) > 0:
                    output.write(article_contents+"\n")
                    article_contents = ""
        content_line = file.readline()
    output.close()
 
'''
将维基百科语料库进行分类
'''
def generate_corpus():
    zhwiki_path = "C:\\Users\\cam\\NLP\\NLP_Week2\\wikiextractor\\extracted\\AA"
    save_path = "C:\\Users\\cam\\NLP\\NLP_Week2\\wikiextractor\\extracted\\AA"
    for i in range(3):
        file_path = os.path.join(zhwiki_path,str("wiki.zh0%s.txt"%str(i)))
        parse_zhwiki(file_path,os.path.join(save_path,"wiki_corpus0%s"%str(i)))
```

## 下面利用课堂讲授的 2-Gram 的办法建立语言模型（库只选取了中文维基的一部分）

```python
# 我们首先要算交事件的概率，这样就把单个的字符两个两个粘在一起。
valid_tokens_str = [str(t) for t in TEXT[:81525738]]
all_2_grams_words = [''.join(valid_tokens_str[i:i+2]) for i in range(len(valid_tokens_str[:-2]))]

# 所有 2-Gram 的总数
_2_gram_sum = len(all_2_grams_words)

#计算词频
_2_gram_counter = Counter(all_2_grams_words)
```

```python
import jieba

# 定义一个拆分字符串的函数 cut()，注意 jieba.cut() 产生的是生成器不能直接使用。

def cut(string): 
    return list(jieba.cut(string))


frequences_all = [f for w, f in words_count.most_common()]
frequences_sum = sum(frequences_all)


def get_prob(word): 
    esp = 1 / frequences_sum
    if word in words_count: 
        return words_count[word] / frequences_sum
    else:
        return esp

    
# 这个函数是为了计算两个词语连在一起的概率，如果在词库中就用频率代替，否自就用“基本事件”的频率代替。
def get_combination_prob(w1, w2):
    if w1 + w2 in _2_gram_counter: return _2_gram_counter[w1 + w2] / _2_gram_sum
    else:
        return 1 / _2_gram_sum

# 利用条件概率公式
def get_prob_2_gram(w1, w2):
    return get_combination_prob(w1, w2) / get_prob(w1)


def langauge_model_of_2_gram(sentence):
    sentence_probability = 1
    
    words = cut(sentence)
    
    # enumerate() 函数用于将一个可遍历的数据对象(如列表、元组或字符串)组合为一个索引序列，同时列出数据和数据下标，一般用在 for 循环当中。
    for i, word in enumerate(words):
        if i == 0: 
            prob = get_prob(word)
        else:
            previous = words[i-1]
            prob = get_prob_2_gram(previous, word)
        sentence_probability *= prob
    
    return sentence_probability
```

如果 w1 没在词库中出现，则 

get_prob(w1) = $\frac{1}{\text{frequences_sum}}$，从而

这意味着

get_prob_2_gram = $\frac{\text{frequences_sum}}{\text{_2_gram_sum}}>1$.

仍用课堂中的例子，我们看到可以进行有效的区分。

```python
need_compared = [
    "今天晚上请你吃大餐，我们一起吃日料 明天晚上请你吃大餐，我们一起吃苹果",
    "真事一只好看的小猫 真是一只好看的小猫",
    "今晚我去吃火锅 今晚火锅去吃我",
]

for s in need_compared:
    s1, s2 = s.split()
    p1, p2 = langauge_model_of_2_gram(s1), langauge_model_of_2_gram(s2)
    
    better = s1 if p1 > p2 else s2
    
    print('{} is more possible'.format(better))
    print('-'*4 + ' {} with probility {}'.format(s1, p1))
    print('-'*4 + ' {} with probility {}'.format(s2, p2))
```

## 下面我们尝试 3-Gram 的办法建立语言模型。
理论依据：对于词语 “$ABC$”，
$$P(C|AB) = \frac{P(ABC)}{P(AB)}.$$
概率用频率近似，则有：
$$P(C|AB) = \frac{n(ABC)}{n(AB)}.$$

```python
# 我们首先要算交事件的概率，这样就把单个的字符三个三个粘在一起。
valid_tokens_str = [str(t) for t in TEXT[:81525738]]
all_3_grams_words = [''.join(valid_tokens_str[i:i + 3]) for i in range(len(valid_tokens_str[:-3]))]

# 所有 3-Gram 的总数
_3_gram_sum = len(all_3_grams_words)

# 计算词频
_3_gram_counter = Counter(all_3_grams_words)
```

```python
import jieba

# 定义一个拆分字符串的函数 cut()，注意 jieba.cut() 产生的是生成器不能直接使用。

def cut(string): 
    return list(jieba.cut(string))


frequences_all = [f for w, f in words_count.most_common()]
frequences_sum = sum(frequences_all)


def get_prob(word): 
    esp = 1 / frequences_sum
    if word in words_count: 
        return words_count[word] / frequences_sum
    else:
        return esp

# 这个函数是为了计算三个词语连在一起的概率，如果在词库中就用频率代替，否自就用“基本事件”的频率代替。
def get_combination_3_prob(w1, w2, w3):
    if w1 + w2 + w3 in _3_gram_counter: return _3_gram_counter[w1 + w2 + w3] / _3_gram_sum
    else:
        return 1 / _3_gram_sum
    
    
# 利用条件概率公式，注意这里依赖 2-gram 的函数
def get_prob_3_gram(w1, w2, w3):
    return get_combination_3_prob(w1, w2, w3) / get_combination_prob(w1, w2)


def langauge_model_of_3_gram(sentence):
    sentence_probability = 1
    
    words = cut(sentence)
    
    # enumerate() 函数用于将一个可遍历的数据对象(如列表、元组或字符串)组合为一个索引序列，同时列出数据和数据下标，一般用在 for 循环当中。
    for i, word in enumerate(words):
        if i == 0: 
            prob = get_prob(word)
        elif i == 1:
            prob = get_prob_2_gram(words[0], word)
        else:
            previous_1 = words[i - 2]
            previous_2 = words[i - 1]
            prob = get_prob_3_gram(previous_1, previous_2, word)
        sentence_probability *= prob
    
    return sentence_probability
```

2. 根据概率，随机选择一个 bigram $(w,x)$;
3. 依次下去，知道选到 $\langle/s\rangle$;
4. 把上述字符串组合起来。
